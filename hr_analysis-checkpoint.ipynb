```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load only first 500 rows
data = pd.read_csv("HR_Data_MNC_Data Science Lovers.csv").head(500)

# Show first 5 rows
print(data.head())

# Check shape
print("Dataset shape:", data.shape)
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Department', hue='Department', data=data, palette='viridis', legend=False)
plt.xticks(rotation=45)
plt.title('Number of Employees by Department (500 sample)')
plt.savefig("1_Employees_by_Department.png")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data['Salary_INR'], bins=30, kde=True, color='blue')
plt.title('Salary Distribution of Employees')
plt.xlabel('Salary (INR)')
plt.ylabel('Count')
plt.savefig("2_Salary_Distribution_of_Employees.png")
plt.show()
```
